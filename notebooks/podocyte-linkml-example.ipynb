# Podocyte LinkML Example

````python
import yaml
import json, jsonschema
from linkml.generators.jsonschemagen import JsonSchemaGenerator
````

## load podocyte example data and jsonschema created by linkml

````python
with open('../src/data/examples/podocyte-001.yaml') as fh:
    podocyte_data = yaml.load(fh, Loader=yaml.SafeLoader)
````

````python
json_schema = json.loads(JsonSchemaGenerator('../src/linkml/cellcard.yaml', top_class="CellCard").serialize())
json_schema["additionalProperties"] = False # don't permit additional properties
````

## validate the data
No output means it passed validation.

````python
jsonschema.validate(instance=podocyte_data, schema=json_schema)
````

## test some invalid data

````python
invalid_json = podocyte_data.copy()
````

### make OBO ID invalid
The schema defines the `obo id` attribute as:
```
obo id:
  range: string
  pattern: ^CL_[0-9]{7}$
  description: >-
    he namepace and identifer portion of an OBO ontology class (e.g., CL_0000000).
 ```
 This specifies that the `obo id` must start with "CL_" and be followed by 7 digits.

````python
invalid_json["obo_id"]
````

### set `obo id` to something invalid

````python
invalid_json["obo_id"] = "CL_00006530"
````

### test validation

````python
jsonschema.validate(instance=invalid_json, schema=json_schema)
````

### reset the `obo id` and test additional properties

````python
invalid_json["obo_id"] = "CL_0000653" # restet obo id to a valid pattern
````

````python
invalid_json["foo"] = "bar" # adding property
````

````python
jsonschema.validate(instance=invalid_json, schema=json_schema)
````
